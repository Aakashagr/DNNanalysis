```python
import xarray as xr
import scipy
import numpy as np
from matplotlib import pyplot as plt
import string
from sklearn.linear_model import LassoCV

data = xr.open_dataarray('assy_dicarlo_Rajalingham2020.nc')

monkeyid = np.zeros(510)
for i,mid in enumerate(np.unique(data['monkey'])):
    monkeyid[data['monkey'] == mid] = i

isword = data['label_word'][::33]
hem = np.array(data['hemisphere'])
```

```python
#  Linear model fit using LASSO
from sklearn.linear_model import LinearRegression

strlist = list(string.ascii_lowercase) # storing charaacters from A-Z
stimword = np.unique(data['word'])
Xmat = np.zeros((len(stimword), 26*4))

# Assigning ones at corresponding locations      
for i, seq in enumerate(stimword):
    for j, char in enumerate(seq): 
        if char in strlist:
            pid = (strlist.index(char)*4) + j
            Xmat[i,pid] += 1 

splithalfr = np.zeros(np.size(data,1))
rfit = np.zeros(np.size(data,1))
modelCoef = np.zeros(np.shape(Xmat))

for neu in range(np.size(data,1)):
    neuresp = np.reshape(data[:,neu],[616,33])
    avgresp_tr = np.nanmean(neuresp[:,::2],1) # Training on odd-repeats and testing on even repeats
    avgresp_ts = np.nanmean(neuresp[:,1::2],1)

    # Calculating odd-even split correlation
    splithalfr[neu],_ = scipy.stats.pearsonr(np.nanmean(neuresp[:,::2],1),np.nanmean(neuresp[:,1::2],1)) 
    
#     Performing cross-validated lasso regression
    reg = LassoCV(cv=5).fit(Xmat, avgresp_tr)  
    reg = LinearRegression().fit(Xmat, avgresp_tr)  # Performing linear regression. 

        
    rfit[neu] = reg.score(Xmat, avgresp_ts)  
    modelCoef[neu,:] = reg.coef_
    print('neuron no. = ' + str(neu))
    
    
```

```python
# Non-linear model fit

#  Training a non-linear regression model on words and testing it on other categories
from scipy.optimize import curve_fit
from scipy import stats

strlist = list(string.ascii_lowercase) # storing charaacters from A-Z
stimword = np.unique(data['word'])
Xmat = np.zeros((len(stimword), 26*4))

# Assigning ones at corresponding locations      
for i, seq in enumerate(stimword):
    for j, char in enumerate(seq): 
        if char in strlist:
            pid = (strlist.index(char)*4) + j
            Xmat[i,pid] += 1 
            
            
def func(x, *coef):
    out = np.zeros((np.shape(x)[0],26))
    for i in range(26):
        out[:,i] = (x[:,i*4:(i+1)*4]@coef[26:-1])*coef[i]
    return np.sum(out,1)+coef[-1]


# Initializing the variables
rfitnl = np.zeros(np.size(data,1))
modelCoefnl = np.zeros((510,31))
splithalfr = np.zeros(np.size(data,1))

for neu in range(np.size(data,1)):
    neuresp = np.reshape(data[:,neu],[616,33])
    avgresp_tr = np.nanmean(neuresp[:,::2],1) # Training on odd-repeats and testing on even repeats
    avgresp_ts = np.nanmean(neuresp[:,1::2],1)
    splithalfr[neu],_ = scipy.stats.spearmanr(np.nanmean(neuresp[:,::2],1),np.nanmean(neuresp[:,1::2],1)) 

    avgresp_tr = avgresp_tr - np.mean(avgresp_tr); avgresp_ts = avgresp_ts - np.mean(avgresp_ts)
    
    # Performing cross-validated lasso regression
    try: 
        popt, _ = curve_fit(func, Xmat, avgresp_tr, p0 = np.ones(31),bounds=(-50, 50))
        rfitnl[neu],_ = stats.spearmanr(avgresp_ts, func(Xmat,*popt))
        modelCoefnl[neu,:] = popt
        print('neuron no. = ' + str(neu))

    except RuntimeError:
        print("Error - curve_fit failed")
```

```python
# csplithalfr2 = (splithalfr*2/(1+splithalfr))**2 # Spearman-brown correction
csplithalfr2 = splithalfr**2 # Uncorrected Spearman-brown correction


plt.scatter( csplithalfr2,rfitnl**2, c = monkeyid)
plt.xlabel('Linear Model fit (R*2)')
# plt.xlabel('Split-half correlation (R*2)')
plt.ylabel('Non-linear model fit (R^2)')
plt.plot([0, 1], [0, 1] , linestyle='-', color = 'k')  
```

```python
# Fitting bigram model
bigrams = []
for lword in stimword:
    for i in range(len(lword)-1):
        bigrams.append(lword[i:i+2])
        
bigramlist =  np.unique(bigrams); bigramlist = list(bigramlist[3:])   

# Building regression matrix
Xmat = np.zeros((len(stimword), len(bigramlist))) 

# Assigning ones at corresponding locations      
for i, seq in enumerate(stimword):
    for j in range(len(seq)-1): 
        char = seq.lower()[j:j+2]
        if char in bigramlist:
            pid = (bigramlist.index(char)) 
            Xmat[i,pid] += 1 
            
# Fitting models of neurons with poor letter model fit
outneu = np.where(rfit +.4 < csplithalfr2)[0]

Osplithalfr = np.zeros(np.size(outneu))
Orfit = np.zeros(np.size(outneu))
OmodelCoef = np.zeros(np.shape(Xmat))


for i,neu in enumerate(outneu):
    neuresp = np.reshape(data[:,neu],[616,33])
    avgresp = np.nanmean(neuresp,1)

    # Calculating odd-even split correlation
    Osplithalfr[i],_ = scipy.stats.pearsonr(np.nanmean(neuresp[:,::2],1),np.nanmean(neuresp[:,1::2],1)) 
    
    # Performing cross-validated lasso regression
    reg = LassoCV(cv=5, random_state=0,max_iter=2000).fit(Xmat, avgresp)   
    Orfit[i] = reg.score(Xmat, avgresp)  
    OmodelCoef[i,:] = reg.coef_
    print('neuron no. = ' + str(neu))
    
    

# np.where(rfit +.4 < csplithalfr2)
# rfit[19]
 
```
