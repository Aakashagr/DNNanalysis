```python
from torchvision import datasets, transforms

# data_dir = 'images/'
data_dir = 'stimuli/vinckier50/'
transform = {'train': transforms.Compose([transforms.Resize((224,224)),
transforms.ToTensor(),
transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])]),}

chosen_datasets = {x: datasets.ImageFolder(os.path.join(data_dir, x), transform[x]) for x in ['train']}
dataloaders = {x: torch.utils.data.DataLoader(chosen_datasets[x], batch_size = 400) for x in ['train']}
images, classes = next(iter(dataloaders['train']))

np.shape(images)
```

```python
stimtemp = images[:,:,:,:]

# Loading all the three network and extracting representation from all the hidden units
ili = {}
net = CORnet_Z_tweak()
checkpoint = torch.load('models/save_illit_z_79_full_nomir.pth.tar')['state_dict']
for key in list(checkpoint.keys()):
    if 'module.' in key:
        checkpoint[key.replace('module.', '')] = checkpoint[key]
        del checkpoint[key]
net.load_state_dict(checkpoint)
# ili['v1'], ili['v2'], ili['v4'], ili['it'], ili['h'],  ili['out'] = net(stimtemp.float())
ili['v1'], ili['v2'], ili['v4'], ili['it'], ili['h'],  ili['out'] = net(stimtemp.float())

Bli = {}
net = CORNet_Z_biased_words()
checkpoint = torch.load('models/save_lit_bias_z_79_full_nomir.pth.tar')['state_dict']
for key in list(checkpoint.keys()):
    if 'module.' in key:
        checkpoint[key.replace('module.', '')] = checkpoint[key]
        del checkpoint[key]
net.load_state_dict(checkpoint)
Bli['v1'], Bli['v2'], Bli['v4'], Bli['it'], Bli['h'],  Bli['out'] = net(stimtemp.float())


nBli = {}
net = CORNet_Z_nonbiased_words()
checkpoint = torch.load('models/save_lit_no_bias_z_79_full_nomir.pth.tar')['state_dict']
for key in list(checkpoint.keys()):
    if 'module.' in key:
        checkpoint[key.replace('module.', '')] = checkpoint[key]
        del checkpoint[key]
net.load_state_dict(checkpoint)

nBli['v1'], nBli['v2'], nBli['v4'], nBli['it'], nBli['h'],  nBli['out'] = net(stimtemp.float())
```

```python
# Correlation between string representation within a given network and layer

from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import pdist, squareform
import pandas as pd
from scipy.special import comb

fig, axs = plt.subplots(2,3, figsize=(12,8), facecolor='w', edgecolor='k')
# fig.subplots_adjust(hspace = 0, wspace=1)
axs = axs.ravel()
qtemp = []
roi_ID = ['v1', 'v2', 'v4','it','h','out']; distMetric = 'correlation'

D = np.zeros([300,300,6])
for idx in range(len(roi_ID)):
    D[:,:,idx] = pairwise_distances(ili[roi_ID[idx]].detach().numpy(),metric = distMetric)
    
for idx in range(len(roi_ID)):
    im = axs[idx].imshow(D[:,:,idx],vmin=0, vmax= max(D.flatten())-.2)
#     axs[idx].set_yticks(np.r_[0, np.arange(80,381,50)]);     axs[idx].set_xticks(np.r_[0, np.arange(80,381,50)])
    axs[idx].set_yticks(np.r_[np.arange(0,301,50)]);     axs[idx].set_xticks(np.r_[np.arange(0,301,50)])
    axs[idx].set_yticklabels(['FF','IL','FL','BG','QG','W'])
    axs[idx].set_xticklabels(['FF','IL','FL','BG','QG','W'])
    axs[idx].set_title(roi_ID[idx])
    qtemp.append(np.mean(D))
    
fig.subplots_adjust(right=0.8)
cbar_ax = fig.add_axes([0.85, 0.15, 0.05, 0.7])
fig.colorbar(im, cax=cbar_ax)
fig.suptitle('Illiterate n/w RDM - Correlation distance metric', fontsize=20)
plt.show()

```

```python
vincdis = loadmat('vindis.mat')['pred_dis']
distMetric = 'correlation'
rili = []; rnbli = []; rbli = []
for idx in range(len(roi_ID)):
    D = pairwise_distances(ili[roi_ID[idx]].detach().numpy()[50:,:],metric = distMetric)
    rili.append(np.corrcoef(squareform(D),vincdis.flatten())[1,0])
    D = pairwise_distances(nBli[roi_ID[idx]].detach().numpy()[50:,:],metric = distMetric)
    rnbli.append(np.corrcoef(squareform(D),vincdis.flatten())[1,0])
    D = pairwise_distances(Bli[roi_ID[idx]].detach().numpy()[50:,:],metric = distMetric)
    rbli.append(np.corrcoef(squareform(D),vincdis.flatten())[1,0])
# np.shape(ili[roi_ID[idx]].detach().numpy())

```

```python

plt.plot(rili); plt.plot(rnbli); plt.plot(rbli); 
plt.xticks(ticks = np.arange(0,6), labels = roi_ID)
plt.legend(['Illiterate','Non-biased literate','Biased literate'])
plt.title('Match to predicted behaviour dissimilarity ')
# np.shape(vincdis)
# B = np.reshape(squareform(D), (-1, 2))
# np.shape(vincdis.flatten())
```

```python
# PCA in python without any normalization or whiten matches PCA of MATLAB
from sklearn.decomposition import PCA

threeD = 0
if not threeD:
    fig, axs = plt.subplots(2,3, figsize=(12,8), facecolor='w', edgecolor='k')
    fig.subplots_adjust(hspace = .3, wspace=.3)
    axs = axs.ravel()
    npc = 2
else: 
    fig = plt.figure(figsize=(12,8))
    npc = 3

for idx in range(len(roi_ID)):
    x = ili[roi_ID[idx]].detach().numpy()
    pcax = PCA(n_components=npc, whiten = 0)
    out = pcax.fit_transform(x)
    
    if threeD:
        axs = fig.add_subplot(2,3,idx+1, projection='3d')
        q = np.array(range(0,50)); 
        colour = ['r','g','b','k','y','c','m']
        for i in range(6):
            axs.scatter3D(out[q, 0], out[q, 1], out[q, 2],marker = 'o',color = colour[i])
            axs.set_title(roi_ID[idx])
            q = q+50;
    else:
        q = np.array(range(0,50)); 
        colour = ['r','g','b','k','y','c','m']
        for i in range(6):
            axs[idx].scatter(out[q, 0], out[q, 1], marker = 'o',color = colour[i])
            axs[idx].set_title(roi_ID[idx])
            q = q+50;

    
LEG = ['FF','IL','FL','BG','QG','W']
plt.legend((LEG[:]), bbox_to_anchor=(1, 1))
fig.suptitle('Illiterate network - PC space', fontsize=20)
```

```python
# Loading stimulus names
import glob, os
import string
wordlist = []; nstimpercat = 50
for file in glob.glob("../../StimulusProperties/*.txt"):
    if not file[-6:-4] == 'NV':
        file1 = open(file,"r") 
        x = (file1.read()).split('\n')
        wordlist = np.concatenate((wordlist, x[:nstimpercat]))

        
# Performing PCA and projecting the data
from sklearn.decomposition import PCA
pcax = PCA(n_components=10)
x = nBli['out'].detach().numpy(); out = pcax.fit_transform(x)
# PCs = np.load('WordPCspace_literate.npy')
# out = nBli['out'].detach().numpy()@np.transpose(PCs)

# Adding PC value of each letter at each location and plotting the average value
from  collections import defaultdict
max_len = max(map(len, wordlist))
fig, axs = plt.subplots(1,5, figsize=(12,8), facecolor='w', edgecolor='k')
axs = axs.ravel(); idx =0

for npc in range(5):
    nocc = defaultdict(lambda: [0]*max_len)  
    pcval = defaultdict(lambda: [0]*max_len)  

    for i, seq in enumerate(wordlist):
        for j, char in enumerate(seq): 
            nocc[char][j] += 1
            pcval[char][j] += out[i,npc]

    # Sorting the key values from A-Z
    sortednames=sorted(nocc.keys(), key=lambda x:x.lower()); xx= []
    for i in sortednames:
        xx.append(np.array(pcval[i])/np.array(nocc[i]))

    # Plotting the data
    axs[npc].imshow(xx);plt.set_cmap('bwr') 
    axs[npc].set_yticks(np.arange(0,26));     axs[npc].set_xticks(np.arange(0,6))
    axs[npc].set_yticklabels(list(string.ascii_uppercase));    
    axs[npc].set_xticklabels(list(np.arange(6)+1));
    axs[npc].set_title('PC #: ' + str(npc+1))

fig.suptitle('Non-biased lliterate network - Out layer', fontsize=20)
```
