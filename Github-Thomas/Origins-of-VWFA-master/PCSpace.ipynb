```python
from sklearn.decomposition import PCA

from matplotlib.pyplot import figure
figure(num=None, figsize=(8, 4), dpi=100, facecolor='w', edgecolor='k')

npc = 100
pcax = PCA(n_components=npc,whiten = 0)
x = nBli['out'].detach().numpy()
out = pcax.fit_transform(x)

# Plotting variance explained
plt.plot(np.cumsum((pcax.explained_variance_ratio_)))
xticklabel =  ["%d" %(i+1) for i in range(0,npc,3)]
plt.xticks(np.r_[:npc:3],xticklabel,rotation=45,size = 8); plt.axhline(.8, color='k', linestyle='--')
plt.xlabel('PC number'); plt.ylabel('Explained variance')
plt.title('out layer')
```
