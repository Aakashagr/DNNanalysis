```python
from sklearn.utils import resample
from scipy.stats import spearmanr

roi_ID = ['v1', 'v2', 'v4','it','h','out']; distMetric = 'euclidean'
Dcorr = np.zeros((6))
for idx in range(len(roi_ID)):
    D = np.zeros((comb(np.size(stimtemp[::2,:],0), 2,exact =True),2)); 

    D[:,0] = squareform(pairwise_distances(nBli[roi_ID[idx]].detach().numpy()[::2,:],metric = distMetric))
    D[:,1] = squareform(pairwise_distances(nBli[roi_ID[idx]].detach().numpy()[1::2,:],metric = distMetric))
    

    Dcorr[idx] = spearmanr(D[:,0],D[:,1])
    
    
    
```

```python
from sklearn.utils import resample
n_iterations = 1000
# run bootstrap
bstrpD = []
for i in range(n_iterations):
    # prepare train and test sets
    Dsampled = resample(D, n_samples=np.size(D,0))
    bstrpD.append(1-(pairwise_distances(Dsampled.T,metric = 'correlation')))
    
```
